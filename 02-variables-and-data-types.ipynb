Every program you write needs to store and manipulate information — a user's name, a price, a yes/no answer. Variables are how Python holds that information in memory. Data types are how Python knows what kind of information it is and what operations make sense on it. This notebook covers both from the ground up.

## What is a Variable?

A variable is a **named label attached to a value**. Instead of writing the same value over and over throughout your program, you give it a name once and refer to it by that name wherever you need it.

Think of it like a sticky note. You write a value on the note, give the note a label, and stick it somewhere you can find it. When you need that value, you look up the label.

In Python, you create a variable with the **assignment operator** `=`:

```
name = value
```

The `=` here does **not** mean equality (as in maths). It means: *take the value on the right, and attach it to the name on the left*.

````python
# Assigning values to variables
city = "Tokyo"
population = 13_960_000   # underscores in numbers are ignored — they just aid readability
area_km2 = 2_194.0

print(city)
print(population)
print(area_km2)
````

### How Python Stores Variables in Memory

The code above creates three names pointing to three objects on the heap. Python never stores a value directly inside a variable name — a name is a **reference** (a pointer), and the object it points to carries both the value and the type.

![Python Variable Memory Model](https://raw.githubusercontent.com/schemabotview/python/main/img/python-variable-memory-model.svg)

Each object on the heap stores three things: its **type** (how Python knows what operations are valid on it), its **value**, and a **reference count** (how many names currently point to it). When the count drops to zero the garbage collector reclaims the memory automatically.

`alias = population` is the key example: assignment never copies a value — it makes a second name point to the **same object**. That is why the integer object shows `refs: 2`. If you later do `alias = 999`, only the pointer for `alias` moves; `population` still points to `13_960_000`.

````python
# Variables can be reassigned — the label moves to the new value
score = 0
print("Initial score:", score)

score = 10
print("After first round:", score)

score = score + 5   # read the current value, add 5, assign back
print("After second round:", score)
````

## Naming Rules & Conventions

Python enforces a few hard **rules** on variable names — break them and your code will not run:

| Rule | Valid | Invalid |
|------|-------|---------|
| Must start with a letter or underscore | `name`, `_count` | `1name`, `@value` |
| Can only contain letters, digits, and underscores | `user_id`, `item2` | `user-id`, `item 2` |
| Case-sensitive | `score` and `Score` are different variables | — |
| Cannot be a Python keyword | `count`, `data` | `for`, `if`, `class`, `return` |

Beyond the rules, Python has a strong **convention**: use **snake_case** — all lowercase letters with underscores between words. This is what the community expects and what you will see in every professional Python codebase.

```python
# Good — snake_case, descriptive
user_age = 28
total_price = 49.99
is_logged_in = True

# Bad — cryptic, hard to read
x = 28
tp = 49.99
flag = True
```

Name variables for what they represent, not how they are stored.

## Dynamic Typing

In some languages (Java, C++), you must declare the type of a variable before you use it:

```java
// Java — type must be declared
int age = 30;
String name = "Alice";
```

Python is **dynamically typed** — you just assign a value, and Python figures out the type for you. You can even reassign a variable to a completely different type later (though that is rarely a good idea in practice).

The built-in `type()` function tells you exactly what type a variable holds at any moment.

````python
# Python infers the type from the value you assign
x = 42
print(x, type(x))

x = 3.14
print(x, type(x))

x = "hello"
print(x, type(x))

x = True
print(x, type(x))
````

## Core Data Types

Python has five built-in types you will use in virtually every program:

| Type | Python name | Example values |
|------|-------------|----------------|
| Integer | `int` | `0`, `42`, `-7`, `1_000_000` |
| Floating-point | `float` | `3.14`, `-0.5`, `2.0`, `1e6` |
| String | `str` | `"hello"`, `'world'`, `"it's fine"` |
| Boolean | `bool` | `True`, `False` |
| Nothing | `NoneType` | `None` |

Let's look at each one in depth.

## Integers — `int`

Integers are **whole numbers** — no decimal point. They can be positive, negative, or zero. Python's `int` has **no fixed size limit** — it will happily handle numbers with hundreds of digits (unlike most other languages, which cap at 64-bit).

````python
age = 28
temperature = -15
distance_to_moon_km = 384_400

# Python integers can be arbitrarily large
googol = 10 ** 100
print(googol)
````

````python
# Integer arithmetic
print(10 + 3)   # addition        → 13
print(10 - 3)   # subtraction     → 7
print(10 * 3)   # multiplication  → 30
print(10 ** 3)  # exponentiation  → 1000
print(10 // 3)  # floor division  → 3  (rounds down, result is int)
print(10 % 3)   # modulo          → 1  (remainder after division)
````

## Floats — `float`

Floats are **numbers with a decimal point**. Python stores them using IEEE 754 double-precision, which gives you about 15–17 significant digits.

One thing every programmer eventually runs into: **floating-point imprecision**. Because computers represent decimals in binary (base-2), some fractions cannot be expressed exactly — just like `1/3` cannot be written exactly in decimal. This is not a Python bug; it is fundamental to how all modern CPUs work.

````python
price = 19.99
pi = 3.14159
tiny = 1e-6     # scientific notation: 0.000001
large = 2.5e9   # scientific notation: 2_500_000_000.0

print(price, type(price))
print(tiny, large)
````

````python
# Regular division always returns a float — even when the result is a whole number
print(10 / 2)   # → 5.0, not 5
print(7 / 2)    # → 3.5

# Floating-point imprecision — this is expected behaviour
print(0.1 + 0.2)          # → 0.30000000000000004, not 0.3
print(round(0.1 + 0.2, 2)) # round() to 2 decimal places gives 0.3
````

## Strings — `str`

Strings are **sequences of text characters** — letters, digits, spaces, punctuation, emoji, anything. You define them by wrapping text in either single quotes `'...'` or double quotes `"..."` — Python treats them identically.

Use double quotes when your text contains an apostrophe, or single quotes when your text contains a double quote — this avoids needing to escape characters.

````python
first_name = "Alice"
last_name = 'Nakamura'
greeting = "It's a great day"        # single quote inside double-quoted string — no issue
quote = 'He said "hello" quietly.'   # double quote inside single-quoted string — no issue

# Triple quotes allow multi-line strings
poem = """Roses are red,
Violets are blue,
Python is readable,
And so should you."""

print(greeting)
print(poem)
````

````python
# String operations
language = "python"

print(len(language))           # length — number of characters → 6
print(language.upper())        # uppercase → 'PYTHON'
print(language.capitalize())   # first letter capitalised → 'Python'

# Concatenation — joining strings with +
full_name = "Alice" + " " + "Nakamura"
print(full_name)

# Repetition — repeating a string with *
print("-" * 20)
````

````python
# f-strings — the cleanest way to embed values inside strings (Python 3.6+)
name = "Alice"
age = 28
city = "Tokyo"

# Put f before the opening quote, then use {variable} inside
message = f"My name is {name}, I am {age} years old, and I live in {city}."
print(message)

# Expressions work inside {} too
price = 9.99
quantity = 3
print(f"Total: ${price * quantity:.2f}")  # :.2f formats to 2 decimal places
````

## Booleans — `bool`

A boolean has exactly **two possible values**: `True` or `False`. Note the capital first letter — that is required in Python.

Booleans are what you get back from comparisons and what `if` statements expect. Under the hood, `True` is just `1` and `False` is just `0` — `bool` is actually a subtype of `int` in Python.

````python
is_logged_in = True
has_admin_access = False

print(is_logged_in, type(is_logged_in))

# Comparisons return booleans
print(10 > 5)    # True
print(10 == 5)   # False  — note: == for comparison, = for assignment
print(10 != 5)   # True
print(3 >= 3)    # True
````

````python
# Truthiness — Python evaluates other types as True or False in a boolean context
# Falsy values: 0, 0.0, "", [], {}, None
# Everything else is truthy

print(bool(0))       # False
print(bool(1))       # True
print(bool(-7))      # True  (any non-zero number is truthy)
print(bool(""))      # False (empty string)
print(bool("hi"))    # True
print(bool(None))    # False
````

## None — the Absence of a Value

`None` is Python's way of representing **nothing** — the deliberate absence of a value. It is its own type: `NoneType`, and there is only ever one `None` object in any Python program.

You will encounter `None` frequently:
- A function that does not explicitly return a value returns `None`
- Variables are sometimes initialised to `None` as a placeholder before they have a real value
- Optional function parameters often default to `None`

The idiomatic way to check for `None` is with `is`, not `==`.

````python
result = None
print(result, type(result))

# Check for None with 'is', not '=='
if result is None:
    print("No result yet")

# Functions that don't return a value implicitly return None
def do_nothing():
    pass

output = do_nothing()
print(output)   # None
````

## Type Conversion (Casting)

Python will not silently convert between types for you — if you try to add a number to a string, you get an error, not a guess. When you need to change a value from one type to another, you do it explicitly using the built-in conversion functions:

| Function | Does |
|----------|------|
| `int(x)` | Converts to integer (truncates floats, parses digit strings) |
| `float(x)` | Converts to float |
| `str(x)` | Converts to string representation |
| `bool(x)` | Converts to boolean using truthiness rules |

This explicitness is intentional — it forces you to be clear about what you want.

````python
# String → int (common when reading user input or parsing data)
user_input = "42"
number = int(user_input)
print(number + 8)   # 50 — now we can do arithmetic

# Float → int (truncates — does NOT round)
print(int(9.9))    # 9, not 10
print(int(-3.7))   # -3, not -4

# int → float
print(float(5))    # 5.0

# Anything → string
print(str(100))    # '100'
print(str(3.14))   # '3.14'
print(str(True))   # 'True'
````

````python
# What happens when conversion is impossible — Python raises an error clearly
int("hello")   # ValueError: invalid literal for int() with base 10: 'hello'
````

````python
# A practical example: reading a number from input()
# input() always returns a string — you must convert it
raw = input("Enter your age: ")
age = int(raw)
print(f"In ten years you will be {age + 10} years old.")
````

## Constants — Values That Should Not Change

Python has no built-in `const` keyword. Instead, the community uses a naming convention: **UPPER_SNAKE_CASE** signals to any reader that this value is intended to be a constant and should not be reassigned.

```python
MAX_RETRIES = 3
PI = 3.14159265358979
DB_HOST = "localhost"
```

Python will not stop you from reassigning them — the capital letters are a social contract, not a technical lock. Honour the convention.

````python
# Constants defined at the top of a file / module
GRAVITY_MS2 = 9.81          # metres per second squared
SPEED_OF_LIGHT_KMS = 299_792  # km/s
MAX_LOGIN_ATTEMPTS = 5

# Use them by name — no magic numbers scattered through the code
distance = 0.5 * GRAVITY_MS2 * (3 ** 2)   # d = ½gt²
print(f"Fell {distance} metres in 3 seconds")
````

## Multiple Assignment

Python lets you assign multiple variables on a single line — useful for conciseness and for swapping values.

````python
# Assign different values to multiple variables at once
x, y, z = 1, 2, 3
print(x, y, z)

# Assign the same value to several variables
a = b = c = 0
print(a, b, c)

# Swapping two variables — no temporary variable needed
first = "Alice"
second = "Bob"

first, second = second, first   # swap in one line
print(first, second)   # Bob Alice
````

## Checking and Inspecting Variables

Two built-in tools are useful when exploring what a variable holds.

````python
price = 19.99
label = "Coffee"
in_stock = True

# type() — tells you the type of a value
print(type(price))     # <class 'float'>
print(type(label))     # <class 'str'>
print(type(in_stock))  # <class 'bool'>

# isinstance() — checks if a value is of a particular type (or a subtype)
print(isinstance(price, float))   # True
print(isinstance(price, int))     # False
print(isinstance(True, int))      # True — bool is a subtype of int
````

## Summary

- A **variable** is a named label attached to a value using `=`. The label can be reassigned at any time.
- Variable names must start with a letter or underscore, contain only letters/digits/underscores, and cannot be Python keywords. Use **snake_case** by convention.
- Python is **dynamically typed** — the type is inferred from the value, not declared upfront. Use `type()` to inspect it.
- The five core built-in types: **`int`** (whole numbers), **`float`** (decimals), **`str`** (text), **`bool`** (`True`/`False`), and **`None`** (no value).
- Float arithmetic can produce tiny imprecision — use `round()` when precision matters.
- Use **f-strings** (`f"...{variable}..."`) to embed values inside strings.
- **Type conversion** is explicit: `int()`, `float()`, `str()`, `bool()`. `int()` truncates, not rounds.
- Use **UPPER_SNAKE_CASE** for values intended to be constants.
- Python supports **multiple assignment** and clean one-line variable swaps.
