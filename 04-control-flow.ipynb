By default, Python runs your code top to bottom, one line at a time. Control flow is how you break out of that straight line — branching to run different code depending on a condition, repeating a block of code for every item in a collection, or looping until something changes. This notebook covers Python's full set of control flow tools: `if`/`elif`/`else`, `for` loops, `while` loops, `break`, `continue`, `pass`, loop `else` clauses, and the `match` statement introduced in Python 3.10.

## `if`, `elif`, `else`

An `if` statement evaluates a condition. If the condition is truthy, its indented block runs. If not, Python moves on.

```
if condition:
    # runs when condition is truthy
elif other_condition:
    # runs when first was falsy and this is truthy
else:
    # runs when all conditions above were falsy
```

Key points:
- The condition does not need parentheses (unlike C, Java, JavaScript).
- The colon `:` at the end of each clause is required.
- Indentation defines the block — Python has no braces. The standard is **4 spaces**.
- `elif` and `else` are optional. You can have as many `elif` branches as you need.
- Only the **first** matching branch runs — Python stops checking once one is satisfied.

````python
score = 74

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score} → Grade: {grade}")  # Score: 74 → Grade: C
````

````python
# Any truthy/falsy value works as a condition — not just booleans
username = ""

if username:
    print(f"Hello, {username}")
else:
    print("Please enter a username")   # ← runs because empty string is falsy

items = ["apple", "banana"]

if items:
    print(f"You have {len(items)} item(s) in your cart")  # ← runs: non-empty list is truthy
else:
    print("Your cart is empty")
````

## Conditional Expressions (Ternary)

Python has a one-line form of `if`/`else` called a **conditional expression** (often called a ternary operator). It evaluates to one of two values depending on a condition.

```python
value_if_true if condition else value_if_false
```

Use it for simple assignments where you would otherwise need four lines of `if`/`else`. Avoid it when the logic is complex — clarity always wins.

````python
age = 20
status = "adult" if age >= 18 else "minor"
print(status)   # adult

# Useful inline — e.g. inside an f-string
cart_count = 1
print(f"You have {cart_count} item{'s' if cart_count != 1 else ''} in your cart")
# You have 1 item in your cart

cart_count = 3
print(f"You have {cart_count} item{'s' if cart_count != 1 else ''} in your cart")
# You have 3 items in your cart
````

## `for` Loops

A `for` loop iterates over any **iterable** — a sequence you can step through one item at a time. Strings, lists, tuples, sets, dictionaries, ranges, and files are all iterable.

```python
for item in iterable:
    # body runs once per item
```

Python's `for` is a **for-each** loop — it always iterates over the items of a collection, not over an index counter. If you need the index too, use `enumerate()`.

````python
# Iterate over a list
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit.upper())

# Iterate over a string — character by character
for char in "Python":
    print(char, end=" ")   # end=" " replaces the default newline
print()  # newline after
````

````python
# enumerate() — gives you index and value together
languages = ["Python", "Go", "Rust"]

for i, lang in enumerate(languages):
    print(f"{i + 1}. {lang}")

# start= shifts the counter
for i, lang in enumerate(languages, start=1):
    print(f"{i}. {lang}")
````

````python
# zip() — iterate two sequences in parallel
names  = ["Alice", "Bob", "Carol"]
scores = [92, 85, 78]

for name, score in zip(names, scores):
    print(f"{name}: {score}")
````

````python
# Iterating over a dictionary
config = {"host": "localhost", "port": 5432, "debug": True}

# .items() gives (key, value) pairs
for key, value in config.items():
    print(f"  {key} = {value}")
````

## `range()`

`range()` generates a sequence of integers on demand — without creating a list in memory. It is the standard way to loop a fixed number of times or to produce an index sequence.

```python
range(stop)              # 0, 1, 2, …, stop-1
range(start, stop)       # start, start+1, …, stop-1
range(start, stop, step) # start, start+step, …, up to but not including stop
```

The `stop` value is **always excluded** — just like slicing.

````python
# Loop 5 times
for i in range(5):
    print(i, end=" ")   # 0 1 2 3 4
print()

# Count from 1 to 10 inclusive
for i in range(1, 11):
    print(i, end=" ")
print()

# Step by 2 — even numbers
for i in range(0, 20, 2):
    print(i, end=" ")
print()

# Negative step — count down
for i in range(10, 0, -1):
    print(i, end=" ")
print()
````

## `while` Loops

A `while` loop keeps running its body as long as its condition remains truthy. Use it when you do not know in advance how many iterations you need — for example, reading input until the user types "quit", or retrying until a network call succeeds.

```python
while condition:
    # body
```

The condition is checked **before** each iteration. If it is falsy from the start, the body never runs. Make sure the body eventually makes the condition false — otherwise you have an infinite loop.

````python
# Countdown
n = 5
while n > 0:
    print(n, end=" ")
    n -= 1   # this line moves the condition toward False
print("Go!")
````

````python
# Simulating retries with a while loop
import random
random.seed(42)

MAX_RETRIES = 5
attempt = 0
success = False

while attempt < MAX_RETRIES and not success:
    attempt += 1
    # Simulate a call that succeeds ~50% of the time
    success = random.random() > 0.5
    print(f"Attempt {attempt}: {'OK' if success else 'failed'}")

if success:
    print("Done.")
else:
    print("All retries exhausted.")
````

## `break` and `continue`

Two keywords give you fine-grained control inside any loop:

- **`break`** — exits the loop immediately. The remaining iterations are abandoned and execution jumps to the first line after the loop.
- **`continue`** — skips the rest of the current iteration and jumps straight to the next one (re-checking the condition in a `while`, advancing to the next item in a `for`).

Both work in `for` and `while` loops.

````python
# break — stop as soon as we find what we want
emails = ["alice@work.com", "bob@work.com", "admin@work.com", "carol@work.com"]

for email in emails:
    if email.startswith("admin"):
        print(f"Admin account found: {email}")
        break   # no point checking the rest
    print(f"Checked: {email}")
````

````python
# continue — skip certain items, process the rest
numbers = [4, -1, 7, -3, 0, 9, -2, 6]

total = 0
for n in numbers:
    if n <= 0:
        continue   # skip non-positives
    total += n

print(f"Sum of positive numbers: {total}")  # 4 + 7 + 9 + 6 = 26
````

## `else` on Loops

Python's `for` and `while` loops can have an `else` clause — a feature that surprises many people coming from other languages.

The `else` block runs **only if the loop completed normally** — i.e., it was not stopped by a `break`. If a `break` fires, the `else` is skipped.

This is most useful when you are searching through a collection and want to act differently depending on whether you found what you were looking for.

````python
# Search for a prime number in a range
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Loop-else: the else runs only if no break occurred
for n in range(20, 30):
    if is_prime(n):
        print(f"First prime found: {n}")
        break
else:
    print("No prime found in range")
````

````python
# while-else: useful for retry loops
import random
random.seed(7)

attempts = 0
while attempts < 3:
    attempts += 1
    if random.random() > 0.7:   # 30% chance of success
        print(f"Connected on attempt {attempts}")
        break
    print(f"Attempt {attempts} failed")
else:
    print("Could not connect after 3 attempts")
````

## `pass`

`pass` is a no-op — it does nothing. Python requires that every block (after `:`) contains at least one statement, so `pass` is a syntactically valid placeholder for blocks you intend to fill in later, or for blocks that genuinely need to do nothing.

````python
# Placeholder for a function body you haven't written yet
def send_notification(user, message):
    pass   # TODO: implement

# Silently ignore specific values in a loop
for n in range(10):
    if n % 2 == 0:
        pass   # skip even numbers without any action
    else:
        print(n, end=" ")
print()
````

## Nested Loops

You can place a loop inside another loop. The inner loop completes all its iterations for each single iteration of the outer loop. The total number of iterations is the product of both loops' lengths, so be mindful of performance when nesting deeply.

A `break` inside a nested loop only exits the **innermost** loop. To exit multiple levels at once, you typically use a flag variable or restructure into a function and use `return`.

````python
# Multiplication table
for row in range(1, 4):
    for col in range(1, 4):
        print(f"{row * col:3}", end="")   # :3 pads to 3 characters
    print()
````

````python
# Break only exits the innermost loop — use a flag to break both
matrix = [
    [1,  2,  3],
    [4,  0,  6],   # 0 is the target
    [7,  8,  9],
]

found = False
for row_idx, row in enumerate(matrix):
    for col_idx, val in enumerate(row):
        if val == 0:
            found = True
            break   # exits inner loop
    if found:
        break       # now exits outer loop

if found:
    print(f"Found 0 at row {row_idx}, col {col_idx}")
````

## `match` Statement (Python 3.10+)

Python 3.10 introduced **structural pattern matching** via `match`/`case`. It is similar in appearance to a `switch` statement from other languages, but far more powerful — it can match values, types, shapes of data structures, and bind parts of the matched value to variables all at once.

```python
match subject:
    case pattern1:
        ...
    case pattern2:
        ...
    case _:      # wildcard — matches anything (like 'default')
        ...
```

Only the **first** matching case runs. The `_` wildcard at the end acts as a catch-all default.

````python
# Simple value matching — cleaner than a long if/elif chain
def http_status(code):
    match code:
        case 200:
            return "OK"
        case 201:
            return "Created"
        case 400:
            return "Bad Request"
        case 401 | 403:     # OR pattern — matches either value
            return "Auth error"
        case 404:
            return "Not Found"
        case 500:
            return "Internal Server Error"
        case _:
            return "Unknown"

for code in [200, 403, 404, 418]:
    print(f"{code}: {http_status(code)}")
````

````python
# Matching on structure — destructuring built into the pattern
def process_command(command):
    match command.split():
        case ["quit"]:
            print("Exiting...")
        case ["go", direction]:
            print(f"Moving {direction}")
        case ["pick", "up", item]:
            print(f"Picked up: {item}")
        case ["drop", *items]:     # *items captures any number of remaining words
            print(f"Dropped: {', '.join(items)}")
        case _:
            print(f"Unknown command: {command!r}")

process_command("go north")
process_command("pick up sword")
process_command("drop map compass")
process_command("quit")
process_command("fly")
````

## Summary

- **`if`/`elif`/`else`**: branch based on conditions. Only the first truthy branch runs. No parentheses needed; colon and indentation are required. Any truthy/falsy value works as a condition.
- **Conditional expressions**: `value_if_true if condition else value_if_false` — concise one-line branching for simple assignments.
- **`for` loops**: iterate over any iterable (list, string, dict, range, …). Use `enumerate()` for index + value, `zip()` for parallel iteration.
- **`range(start, stop, step)`**: generates integers on demand. `stop` is always excluded.
- **`while` loops**: repeat while a condition is truthy. The condition is checked before each iteration. Ensure something in the body moves toward termination.
- **`break`**: exits the current loop immediately. **`continue`**: skips the rest of the current iteration and moves to the next.
- **Loop `else`**: runs only if the loop finished without hitting a `break` — useful for the "not found" case in a search.
- **`pass`**: a syntactic no-op used as a placeholder for empty blocks.
- **Nested loops**: `break` only exits the innermost loop. Use a flag variable or a function with `return` to escape multiple levels.
- **`match`/`case`** (Python 3.10+): structural pattern matching — cleaner than long `if`/`elif` chains and can destructure sequences and objects in the pattern itself.
