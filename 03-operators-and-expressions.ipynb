Operators are the symbols that tell Python how to compute and combine values. An expression is any combination of values, variables, and operators that Python can evaluate down to a single result. This notebook walks through every category of Python operator — arithmetic, comparison, logical, assignment, identity, membership, and bitwise — and explains operator precedence so you know exactly what Python evaluates first.

## Arithmetic Operators

Arithmetic operators perform mathematical calculations on numbers. Python provides seven of them:

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `+` | Addition | `7 + 3` | `10` |
| `-` | Subtraction | `7 - 3` | `4` |
| `*` | Multiplication | `7 * 3` | `21` |
| `/` | Division | `7 / 2` | `3.5` |
| `//` | Floor division | `7 // 2` | `3` |
| `%` | Modulo | `7 % 2` | `1` |
| `**` | Exponentiation | `2 ** 8` | `256` |

A few things worth knowing:
- `/` always returns a `float`, even when the result is a whole number (`10 / 2` → `5.0`).
- `//` returns the floor — it rounds **down** to the nearest integer, not toward zero. `-7 // 2` is `-4`, not `-3`.
- `%` (modulo) returns the **remainder** after floor division. It is more useful than it looks: `n % 2` tells you if `n` is even or odd; `n % 10` gives you the last digit of `n`.

````python
a, b = 17, 5

print(a + b)   # 22 — addition
print(a - b)   # 12 — subtraction
print(a * b)   # 85 — multiplication
print(a / b)   # 3.4 — true division (always float)
print(a // b)  # 3   — floor division (drops remainder)
print(a % b)   # 2   — modulo (remainder)
print(a ** b)  # 1419857 — exponentiation (17^5)
````

````python
# Practical uses of floor division and modulo
total_seconds = 3725

hours   = total_seconds // 3600
minutes = (total_seconds % 3600) // 60
seconds = total_seconds % 60

print(f"{hours}h {minutes}m {seconds}s")  # 1h 2m 5s

# Check even / odd
for n in [4, 7, 100, 13]:
    parity = "even" if n % 2 == 0 else "odd"
    print(f"{n} is {parity}")
````

## Comparison Operators

Comparison operators compare two values and always return a boolean — `True` or `False`. They are the building blocks of conditions.

| Operator | Meaning |
|----------|---------|
| `==` | Equal to |
| `!=` | Not equal to |
| `<` | Less than |
| `>` | Greater than |
| `<=` | Less than or equal to |
| `>=` | Greater than or equal to |

The most common mistake: using `=` (assignment) instead of `==` (comparison). `score = 10` sets a variable; `score == 10` asks a question.

Python also supports **chained comparisons**, which read exactly like mathematical notation and are more concise than combining two comparisons with `and`.

````python
x = 10

print(x == 10)   # True
print(x != 10)   # False
print(x > 5)     # True
print(x < 5)     # False
print(x >= 10)   # True
print(x <= 9)    # False
````

````python
# Chained comparisons — Pythonic and readable
age = 25

# Instead of: age >= 18 and age <= 65
print(18 <= age <= 65)   # True — reads just like maths

score = 72
print(0 <= score <= 100)  # True — valid score range

# Comparisons work on strings too — alphabetical order
print("apple" < "banana")   # True
print("Z" < "a")            # True — uppercase letters come before lowercase in Unicode
````

## Logical Operators

Logical operators combine boolean values or expressions. Python uses plain English words — `and`, `or`, `not` — rather than symbols like `&&`, `||`, `!`.

| Operator | Returns `True` when… |
|----------|----------------------|
| `and` | **both** operands are truthy |
| `or` | **at least one** operand is truthy |
| `not` | the operand is **falsy** |

**Short-circuit evaluation**: Python stops evaluating as soon as the result is determined. For `and`, if the left side is falsy, the right side is never evaluated. For `or`, if the left side is truthy, the right side is never evaluated. This matters when the right side has side effects or could cause errors.

````python
is_adult = True
has_ticket = False
is_vip = True

print(is_adult and has_ticket)      # False — both must be True
print(is_adult or has_ticket)       # True  — at least one is True
print(not has_ticket)               # True  — flips False to True
print(is_adult and (has_ticket or is_vip))  # True — VIP overrides missing ticket
````

````python
# Short-circuit evaluation — the right side is skipped when not needed
items = []

# Without short-circuit, items[0] would raise an IndexError on an empty list
# With 'and', Python checks len(items) > 0 first — if False, stops immediately
if len(items) > 0 and items[0] == "apple":
    print("First item is apple")
else:
    print("List is empty or first item is not apple")  # ← this runs safely
````

````python
# 'or' as a default-value trick — returns the first truthy value it finds
username = ""
display_name = username or "Guest"   # username is falsy (empty string), so uses "Guest"
print(display_name)   # Guest

username = "alice"
display_name = username or "Guest"
print(display_name)   # alice
````

## Assignment Operators

The basic assignment operator `=` stores a value in a variable. Python also has **augmented assignment operators** that combine an arithmetic operation with assignment — they modify a variable in place and are a concise alternative to writing `x = x + 5`.

| Operator | Equivalent to |
|----------|---------------|
| `x += n` | `x = x + n` |
| `x -= n` | `x = x - n` |
| `x *= n` | `x = x * n` |
| `x /= n` | `x = x / n` |
| `x //= n` | `x = x // n` |
| `x %= n` | `x = x % n` |
| `x **= n` | `x = x ** n` |

````python
score = 0

score += 10    # scored a point
print(score)   # 10

score += 5     # another point
print(score)   # 15

score *= 2     # bonus multiplier
print(score)   # 30

score -= 3     # penalty
print(score)   # 27

score //= 4    # floor division in place
print(score)   # 6
````

## Identity Operators — `is` and `is not`

`==` checks if two values are **equal**. `is` checks if two variables point to the **same object in memory** — not just equal, but literally the same object.

This distinction matters most with `None`. Use `is None` and `is not None` — never `== None`. The reason: a custom class can override `==` to return anything it likes, making `== None` unreliable. `is` cannot be overridden.

Python caches small integers (typically -5 to 256) and interned strings, so `is` can return `True` for them as a CPython implementation detail — but don't rely on it. Use `is` only for `None`, `True`, and `False`.

````python
# == checks value equality; 'is' checks object identity
a = [1, 2, 3]
b = [1, 2, 3]   # same values, different list objects
c = a           # c points to the SAME object as a

print(a == b)   # True  — values are equal
print(a is b)   # False — different objects in memory
print(a is c)   # True  — same object
````

````python
# The correct way to check for None
result = None

if result is None:
    print("No result yet")        # ← correct idiom

if result is not None:
    print("Got a result:", result)
else:
    print("Still nothing")
````

## Membership Operators — `in` and `not in`

Membership operators test whether a value exists inside a collection — a string, list, tuple, set, or dictionary.

- `x in collection` — `True` if `x` is found in `collection`
- `x not in collection` — `True` if `x` is **not** found in `collection`

On dictionaries, `in` checks **keys**, not values.

````python
fruits = ["apple", "banana", "cherry"]

print("banana" in fruits)       # True
print("mango" in fruits)        # False
print("mango" not in fruits)    # True

# Works on strings — checks for substrings
email = "user@example.com"
print("@" in email)             # True
print(".org" in email)          # False

# Works on dicts — checks keys
config = {"debug": True, "port": 8080}
print("port" in config)         # True
print("host" in config)         # False
````

## Bitwise Operators

Bitwise operators work directly on the binary representation of integers — they operate bit by bit. You will encounter them in systems programming, networking, cryptography, and anywhere flags or permissions are packed into a single integer.

| Operator | Name | What it does |
|----------|------|---------------|
| `&` | AND | `1` only where **both** bits are `1` |
| `\|` | OR | `1` where **either** bit is `1` |
| `^` | XOR | `1` where bits are **different** |
| `~` | NOT | Flips all bits (`~x` = `-x - 1`) |
| `<<` | Left shift | Shifts bits left (multiplies by 2ⁿ) |
| `>>` | Right shift | Shifts bits right (divides by 2ⁿ) |

````python
a = 0b1100   # 12 in binary
b = 0b1010   # 10 in binary

print(bin(a & b))   # 0b1000 — AND: only bits both have
print(bin(a | b))   # 0b1110 — OR:  bits either has
print(bin(a ^ b))   # 0b0110 — XOR: bits one has but not both
print(bin(~a))      # -0b1101 — NOT: flips every bit

# Shifts are a fast way to multiply/divide by powers of 2
print(1 << 4)   # 16  — 1 shifted left 4 places = 2^4
print(64 >> 3)  # 8   — 64 shifted right 3 places = 64 / 2^3
````

````python
# Practical example: permission flags packed into a single integer
READ    = 0b001   # 1
WRITE   = 0b010   # 2
EXECUTE = 0b100   # 4

# Give a user read + write permissions
user_perms = READ | WRITE   # 0b011 = 3

# Check if user has execute permission
print(bool(user_perms & EXECUTE))  # False — no execute bit set
print(bool(user_perms & READ))     # True  — read bit is set

# Add execute permission
user_perms |= EXECUTE
print(bool(user_perms & EXECUTE))  # True  — now set
````

## Operator Precedence

When an expression contains multiple operators, Python evaluates them in a specific order — just like BODMAS/PEMDAS in maths. Higher precedence means evaluated first.

From **highest to lowest** precedence (simplified):

| Priority | Operators |
|----------|-----------|
| 1 (highest) | `()` — parentheses |
| 2 | `**` — exponentiation |
| 3 | `+x`, `-x`, `~x` — unary operators |
| 4 | `*`, `/`, `//`, `%` |
| 5 | `+`, `-` |
| 6 | `<<`, `>>` — bitwise shifts |
| 7 | `&` |
| 8 | `^` |
| 9 | `\|` |
| 10 | `==`, `!=`, `<`, `>`, `<=`, `>=`, `is`, `is not`, `in`, `not in` |
| 11 | `not` |
| 12 | `and` |
| 13 (lowest) | `or` |

**Practical advice**: rather than memorising all levels, use parentheses to make your intent explicit. `(a + b) * c` is instantly clear; `a + b * c` requires the reader to recall that `*` beats `+`.

````python
# Precedence in action
print(2 + 3 * 4)       # 14 — * evaluated before +
print((2 + 3) * 4)     # 20 — parentheses override

print(2 ** 3 ** 2)     # 512 — ** is right-associative: 3**2=9, then 2**9=512
print((2 ** 3) ** 2)   # 64  — left-to-right with parentheses

print(10 - 4 / 2)      # 8.0 — / before -, result is float

# Logical operator precedence: not > and > or
print(True or False and False)   # True — 'and' binds tighter: or(True, and(False,False))
print(not True or True)          # True — 'not' binds tighter: or(not True, True) = or(False, True)
````

````python
# When in doubt, use parentheses — clarity beats cleverness
age = 20
has_id = True
is_guest_list = False

# Ambiguous — relies on precedence knowledge
allowed = age >= 18 and has_id or is_guest_list

# Clear — intent is explicit
allowed = (age >= 18 and has_id) or is_guest_list

print(allowed)  # True
````

## Expressions vs Statements

An **expression** is any piece of code that evaluates to a value. You can assign it, print it, or pass it to a function.

```python
3 + 4          # expression → 7
name.upper()   # expression → 'ALICE'
x > 0          # expression → True or False
```

A **statement** is an instruction that *does something* but does not evaluate to a value.

```python
x = 10         # assignment statement
if x > 0: ...  # if statement
for i in ...:  # for statement
```

The distinction matters because you can nest expressions inside other expressions — but you cannot put a statement where Python expects a value. Python 3.8 introduced the **walrus operator** `:=`, which is an assignment expression — it assigns and returns the value at the same time.

````python
# Expressions can be composed — they reduce to a single value
result = (3 + 4) * (10 - 2) / 2 ** 2   # one expression
print(result)   # 14.0

# The walrus operator := — assign and use in one step (Python 3.8+)
import re

text = "Order number: 4821"

# Without walrus: call match() twice or use a temp variable
# With walrus: assign the result and test it in one expression
if match := re.search(r"\d+", text):
    print(f"Found order number: {match.group()}")   # 4821
````

````python
# Walrus in a while loop — read until empty input
data = ["line one", "line two", "", "line four"]
index = 0

while line := (data[index] if index < len(data) else None):
    print("Processing:", line)
    index += 1
````

## Summary

- **Arithmetic operators**: `+`, `-`, `*`, `/` (float), `//` (floor), `%` (remainder), `**` (power). `/` always returns a float; `//` floors toward negative infinity.
- **Comparison operators**: `==`, `!=`, `<`, `>`, `<=`, `>=`. Python supports chained comparisons like `0 <= x < 100`.
- **Logical operators**: `and`, `or`, `not`. Short-circuit — stop evaluating as soon as the result is known. `or` returns the first truthy value.
- **Augmented assignment**: `+=`, `-=`, `*=`, `/=`, etc. Modify a variable in place concisely.
- **Identity operators**: `is` / `is not`. Test object identity, not equality. Use exclusively for `None`, `True`, and `False`.
- **Membership operators**: `in` / `not in`. Test whether a value exists in a string, list, tuple, set, or dict (keys).
- **Bitwise operators**: `&`, `|`, `^`, `~`, `<<`, `>>`. Operate on binary representations — useful for flags and low-level work.
- **Precedence**: `**` > unary > `* / // %` > `+ -` > comparisons > `not` > `and` > `or`. Use parentheses to make intent explicit.
- **Walrus operator** `:=`: assigns a value and returns it as an expression — useful in `if` and `while` conditions to avoid repeating a call.
