## What is Python?

Python is a **general-purpose, high-level programming language** — meaning it was designed to be readable by humans first and computers second.

It was created in the late 1980s by a Dutch programmer named **Guido van Rossum**. Guido wanted a language that was simple, clean, and fun to use — something that felt closer to plain English than to the cryptic symbols of older languages like C. He named it after the British comedy group *Monty Python's Flying Circus*, not the snake.

Python 1.0 was released in **1991**. Today, Python 3 is the standard, and it consistently ranks as the **most popular programming language in the world** — used by beginners learning to code for the first time and by engineers at Google, NASA, and Netflix.

| Year | Milestone |
|------|-----------|
| **1989** | Guido van Rossum begins writing Python over Christmas holiday |
| **1991** | Python 1.0 released publicly |
| **2000** | Python 2.0 released — added list comprehensions, garbage collection |
| **2008** | Python 3.0 released — cleaner, modern, but broke backward compatibility |
| **2020** | Python 2 officially retired (end of life) |
| **2023+** | Python 3.11–3.13 — significant speed improvements, better error messages |

## The Zen of Python

Python has a built-in "philosophy" document called the **Zen of Python**, written by Tim Peters. It describes the design principles that guide the language and the community around it. You can read it any time by running `import this`.

````python
import this
````

A few principles worth understanding as a beginner:

**"Readability counts."**
Code is read far more often than it is written — by your future self, by teammates, by anyone who maintains it. Python rewards code that is easy to skim and understand.

**"Simple is better than complex."**
If you can solve a problem with a straightforward 5-line approach or a clever 1-line approach, the 5-line version is usually the better choice. Cleverness has a cost.

**"Errors should never pass silently."**
When something goes wrong, Python should say so loudly, not swallow the error and move on. Hiding bugs makes them much harder to find later.

**"There should be one — and preferably only one — obvious way to do it."**
Unlike some languages that offer a dozen equally valid paths to the same result, Python tries to have a single idiomatic way. This is why Python code written by different people tends to look similar.

## Python 3 vs Python 2

You may occasionally see old tutorials or Stack Overflow answers that use **Python 2**. Here is everything you need to know:

- Python 2 was officially **retired on January 1, 2020**. It receives no bug fixes, no security patches, nothing.
- Python 3 has been the present and future of the language since 2008.
- The two versions are **not fully compatible** — Python 2 code will often fail to run in Python 3.

The most visible difference you will encounter in old examples is the `print` statement:

```python
# Python 2 — print was a statement, no parentheses
print "Hello, World!"

# Python 3 — print is a function, parentheses required
print("Hello, World!")
```

**Always use Python 3.** If you see a tutorial using `print` without parentheses, it is outdated. When in doubt, check which version a resource was written for.

## Installing Python

Python does not come pre-installed in a usable form on most systems (macOS ships an old version for system use — do not rely on it). Here is how to get a fresh Python 3 install on each platform.

---

### macOS — using Homebrew

[Homebrew](https://brew.sh) is the standard package manager for macOS. Install it first if you don't have it, then:

```bash
# Install Homebrew (paste into Terminal)
/bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)"

# Install Python 3
brew install python

# Verify the install
python3 --version
```

---

### Windows — using python.org

1. Go to [https://www.python.org/downloads/](https://www.python.org/downloads/)
2. Click **Download Python 3.x.x** (the latest stable release)
3. Run the installer — **check the box that says "Add Python to PATH"** before clicking Install
4. Open Command Prompt and verify:

```bash
python --version
```

---

### Linux — using apt (Ubuntu / Debian)

```bash
# Update package lists
sudo apt update

# Install Python 3 and pip
sudo apt install python3 python3-pip python3-venv

# Verify
python3 --version
```

Most modern Ubuntu and Debian systems already have Python 3 — this command ensures you have `pip` and `venv` as well.

## Virtual Environments

A **virtual environment** is an isolated copy of Python that belongs to one project. Think of it like a separate room for each project's tools — the hammer in Room A stays in Room A and never clashes with the saw in Room B.

Without virtual environments, every package you install goes into one global pile. Eventually, Project A needs `requests==2.28` and Project B needs `requests==2.31`, and they cannot both win. Virtual environments solve this completely.

**Create and activate a virtual environment:**

```bash
# Step 1: create the environment (creates a folder called 'venv')
python3 -m venv venv

# Step 2: activate it
# macOS / Linux:
source venv/bin/activate

# Windows:
venv\Scripts\activate
```

Once activated, your terminal prompt will show `(venv)` at the start. Every `pip install` from that point on installs only into this environment — your global Python is untouched.

**To deactivate** (go back to the global Python):

```bash
deactivate
```

> **Convention:** name your virtual environment folder `venv` and add it to `.gitignore`. You never commit the environment itself — only the list of packages needed to recreate it (more on that in the next section).

## pip — Installing Packages

`pip` is Python's package installer. It connects to the **Python Package Index** ([pypi.org](https://pypi.org)) — a public repository of over 500,000 free packages — and downloads whatever you need.

```bash
# Install a single package
pip install requests

# Install a specific version
pip install requests==2.31.0

# Install multiple packages at once
pip install requests pandas numpy

# See everything installed in the current environment
pip freeze

# Save that list to a file (the standard convention)
pip freeze > requirements.txt

# Recreate an environment from a requirements file
pip install -r requirements.txt
```

The `requirements.txt` file is what you commit to version control instead of the `venv` folder itself. Anyone who clones your project runs `pip install -r requirements.txt` and gets an identical environment.

## Running Python — Three Ways

There are three main ways to run Python code. Each suits a different situation.

---

### 1. The REPL (Interactive Shell)

REPL stands for **Read–Eval–Print Loop**. It is a live prompt where you type one line of Python and immediately see the result. Great for quick experiments.

```bash
# Start it in your terminal
python3

# You'll see:
# Python 3.12.0 (main, ...)
# >>> 

>>> 2 + 2
4
>>> print("hello")
hello
>>> exit()   # to quit
```

---

### 2. Running a `.py` Script

For real programs, you write your code in a `.py` file and run it from the terminal. This is the standard way to ship Python programs.

```bash
# Create a file called hello.py, then run it:
python3 hello.py
```

---

### 3. Jupyter Notebook

A Jupyter notebook (like this file) mixes runnable code cells with markdown explanation cells. It is the dominant tool in data science and education because you can run code in small chunks, see output inline, and write notes alongside the code.

```bash
# Install Jupyter
pip install jupyter

# Launch the notebook server
jupyter notebook
```

Your browser opens automatically. From there you can create a new `.ipynb` file and start writing.

## How Python Runs Your Code

When you run `python hello.py`, more happens than you might think. CPython — the standard Python interpreter — takes your source code through several distinct stages before a single instruction executes.

![How CPython Executes Your Code](https://raw.githubusercontent.com/schemabotview/python/main/img/python-execution-model.svg)

- **Source code** — the `.py` file you write. Plain text that a human can read.
- **Parser** — breaks the text into tokens (the *lexer*), then builds an **Abstract Syntax Tree (AST)**: a structured tree that represents what your program means, not how it is spelled.
- **Compiler** — walks the AST and emits **bytecode**: a compact stream of low-level opcodes saved to `__pycache__/` as `.pyc` files. Bytecode is platform-independent — the same `.pyc` runs on any machine with the same Python version.
- **PVM (Python Virtual Machine)** — a tight loop that fetches each opcode, decodes it, and executes it. This is the only stage where real computation happens.

Bytecode is cached so that unchanged files skip parsing and compilation on subsequent runs — only the PVM step runs.

## Hello, World!

By tradition, the first program anyone writes in a new language prints the words "Hello, World!" to the screen. In Python, that is a single line.

````python
# The classic first program — one line is all it takes
print("Hello, World!")
````

````python
# print() can output any text you put inside the quotes
print("My name is Python.")
print("I was born in 1991.")
print("I am 35 years old.")
````

````python
# You can build messages dynamically using f-strings (covered in depth later)
# For now, notice how {name} gets replaced with the actual value
names = ["Alice", "Bob", "Carol"]

for name in names:
    print(f"Hello, {name}!")  # f-string: the f before the quote enables {variable} substitution
````

````python
# print() works with numbers and expressions too
print("2 + 2 =", 2 + 2)              # addition
print("10 / 3 =", 10 / 3)            # division always returns a float in Python 3
print("2 to the power of 8 =", 2**8) # ** is the exponentiation operator
````
